```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
bCancer = load_breast_cancer()
bCancer.keys()
```

```python
print(bCancer['DESCR'])
```

```python
pd_cancer = pd.DataFrame(data = bCancer['data'],
                     columns = bCancer['feature_names'])
```

```python
pd_cancer.head()
```

```python
pd_cancer.describe()
```

```python
pd_cancer.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
skv = StandardScaler()
```

```python
skv.fit(pd_cancer)
```

```python
scaledData = skv.transform(pd_cancer)
```

```python
# PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaledData)
```

```python
x_pca = pca.transform(scaledData)
```

```python
scaledData.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=bCancer['target'])

plt.xlabel('FPC1')
plt.ylabel('FPC2')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=bCancer['feature_names'])
df_comp.head()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_comp)
```
